```python
# Types of Graphs

# 1.LINE GRAPH

# 2.BAR GRAPH
# Two types: Vertical(=>bar()) and horizontal bar graph(=>barh())
# We can use bar() function to draw vertical bar graphs
# we can use barh() function to draw horizontal bar graphs
# First two arguments describe the layout of the bars i.e x and y axis
```

```python
x=np.array(['IND','US','AUS','UK'])
y=np.array([205,250,190,105])
plt.title("Bar graph")
plt.bar(x,y,color='b', width=0.8)#width=1 will join bars together without space
plt.show()
```

```python
# 3. HISTOGRAMS
# Syntax: hist()
# A histogram represents data provided in a form of some groups
# It graphically represents the bin ranges while Y-axis gives information about frequency.
# hist(array,bins)
# Here hist() uses an array of numbers sent as argument to create a histogram
# Bins are optional parameter defining consecutive,non-overlapping intervals of variables.

# Creating dataset
xarr=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27,100,111,150,130])
# Creating histogram
plt.hist(xarr,bins=[0,20,40,70,100,200])
# Show plot
plt.show()
```

```python
# 4.PIE CHART
yarr=np.array([10,30,25,35])
labelsArr=["IND", "US", "UK","AUS"]
explodeArr=[0.2,0,0,0]
plt.pie(yarr,labels=labelsArr, explode=explodeArr)
plt.legend(loc='lower right', title='Countries')
plt.show()
```

```python
# 5. SCATTER PLOT
# It is a diagram where each value in the data set is represented by a dot

x=['Aman','Beena','Chaitanya','Deepak','Elaa']
y=[89,76,97,88,100]
plt.scatter(x,y)
plt.show()
```

```python
import pandas as pd
pstore=pd.read_csv('googleplaystore.csv')
# Handling data types
# errors: {'ignore','raise','coerce'}, default 'raise'
# if raise: then invalid parsing will raise an exception
# if coerce: thenn invalid parsing will be set as NaN
# if ignore: then invalid parsing will return the input
pstore['Reviews'] = pd.to_numeric(pstore['Reviews'],errors='coerce')
```

```python
# Removing '+' and ',' from Installs and converting to numeric
pstore['Installs']=pstore['Installs'].str.replace('+','').str.replace(',','').astype(int)
```

```python
pstore['Price'] = pstore['Price'].str.replace('$', '').astype(float)
```

```python
pstore['LastUpdated'] = pd.to_datetime(pstore['LastUpdated'])
```

```python
# Dropping Duplicate Rows
pstore = pstore.drop_duplicates()
pstore.shape
```

```python
import numpy as np
pstore['Size'] = pstore['Size'].replace('Varies with device', np.nan)
pstore
```

```python
# Remove 'M' and 'k' and convert to float
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1000  # Convert kilobytes to megabytes
    else:
        return np.nan

pstore['Size'] = pstore['Size'].apply(lambda x: convert_size(x) if isinstance(x, str) else x)
pstore
```

```python
import seaborn as sns
print(sns.__version__)
```

```python
# Grouping and aggregating data
# Rounding off the mean values to zero decimal places
category_data = pstore.groupby('Category')[['Rating', 'Reviews', 'Installs']].mean().round(2)
print(category_data)
```

```python
correlation = pstore[['Rating', 'Reviews', 'Size', 'Installs', 'Price']].corr()
correlation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
